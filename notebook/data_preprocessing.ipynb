Create a Function to load and preprocess data

```python
def import_clean_data(path,sheetname):
    '''
        This function imports excel data given path and sheetname
        Drops unwanted columns, null value rows and 
        Transposes the df for further processing, further
        Creates a month column 
    '''
    df = pd.read_excel(path,sheet_name=sheetname)
    df.drop(labels=['KPI TYPE','REMARKS / COMMENTS'],axis=1,inplace=True)
    df.dropna(how='all',inplace=True)
    dfT = df.T
    dfT.columns = dfT.iloc[0]
    dfT = dfT.iloc[1:]
    dfT.reset_index(inplace=True)
    dfT['month'] = [i.split(' ')[0] for i in dfT['index']]
    dfT.drop(columns='index',inplace=True)
    return dfT
```
