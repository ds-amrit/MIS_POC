This notebook contains different plots tried for financial data

```python
# Import required libraries for this notebook
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import  cm
from matplotlib.cm import get_cmap
import numpy as np
```

```python
# Load finance df
df = pd.read_csv(r'data\MIS 2018\finance_df.csv')
df.head()
```

```python
# Filter features to plot in the figure
feat_list = list(df.columns)[1:-1]
print(feat_list)
```

```python
# Define variables for plotting
months = list(df['month'])
feature_list = ['# Cash&Carry Retail Stores','Revenues ($K)','Net Flow (Burn)  ($K)']

```

```python
def myplot(data,featurelist):
    c_dict={"Net Flow (Burn)  ($K)":'red',"# Cash&Carry Retail Stores":'black',"Revenues ($K)":'blue'}
    m_dict= {"Net Flow (Burn)  ($K)":'v',"# Cash&Carry Retail Stores":'+',"Revenues ($K)":'^'}
    for f in featurelist:
        plt.plot(data[f],color=c_dict[f], ls = "--", marker=m_dict[f], ms=6, label=f)
    plt.legend(loc = 'upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(len(months))),months, rotation='vertical')
    plt.show()
myplot(df,feature_list)
```

```python
def my_plot(data,featurelist):  
    for f in featurelist:
        plt.plot(data[f],color='b', ls = "--", marker='o', ms=6, label=f)
    plt.legend(loc = 'upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(len(months))),months, rotation='vertical')
    plt.show()
my_plot(df,feature_list)
```

```python
my_plot(df,feat_list)
```

```python
color = iter(plt.cm.rainbow(np.linspace(0, 1, len(feat_list))))
def my_plot_c(data,featurelist):  
    for f in featurelist:
        c = next(color)
        plt.plot(data[f],color=c, ls = "--", marker='o', ms=6, label=f)
    plt.legend(loc = 'upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(len(months))),months, rotation='vertical')
    plt.show()
my_plot_c(df,feat_list)
```

```python
name = "tab20"
cmap = get_cmap(name)  # type: matplotlib.colors.ListedColormap
colors = iter(cmap.colors)  # type: list
#axes.set_prop_cycle(color=colors)

def my_plot_cmap(data,featurelist):  
    for f in featurelist:
        c = next(colors)
        plt.plot(data[f],color=c, ls = "--", marker='o', ms=6, label=f)
    plt.legend(loc = 'upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(len(months))),months, rotation='vertical')
    plt.show()
my_plot_cmap(df,feat_list)
```
